implementing multithreading in python

```python
import threading
def hello(name):
    print(f'hi {name}!')
def bye(name):
    print(f'bye {name}!')
if __name__ == '__main__':
    # create threads for a particular function
    t1 = threading.Thread(target=hello,args=('yash',))
    t2 = threading.Thread(target=bye,args=('yash',))
    # start thread
    t1.start()
    t2.start()
    t1.join()
    t2.join()
    print('done')
```

```python
import threading
import os
def hello():
    print(f'Name: {threading.current_thread().name}')
    print(f'PID : {os.getpid()}')
def bye():
    print(f'Name: {threading.current_thread().name}')
    print(f'PID : {os.getpid()}')
if __name__ == '__main__':
    # print PID of a process
    print(f'Process PID : {os.getpid()}')
    # print the name of process
    print(f'Process Name : {threading.current_thread().name}')
    # create the threds
    t1 = threading.Thread(target=hello,name='T1')
    t2 = threading.Thread(target=bye,name='T2')
    # start the threads
    t1.start()
    t2.start()
    # will the execution of all threads
    t1.join()
    t2.join()
    print('done!')
```

implementing multi-threading in python

```python
from multiprocessing.synchronize import Lock
import threading
# global var
x = 0
def inc():
    global x
    x += 1
def thread_task():
    for i in range(100000):
        Lock.acquire()
        inc()
        Lock.release()
def fun():
    global x
    x = 10
    # create a lock instance
    lock = threading.Lock()
    # create threads
    t1 = threading.Thread(target=thread_task,args=(lock,))
    t2 = threading.Thread(target=thread_task,args=(lock,))
    t1.start()
    t2.start()
    t1.join()
    t2.join()
if __name__ == '__main__':
    for i in range(10):
        fun()
        print(f'Iteration: {i}, X : {x}')
```
