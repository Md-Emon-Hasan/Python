__defining function__

```python
def fun(x):
    return 2*x+1
print(fun(1))
print(fun(2))
print(fun(999))
```

```python
def new_abs(x):
    if x < 0:
        ans = -x
    else:
        ans = x
    return ans
print(new_abs(-9))
print(new_abs(9))
print(new_abs(7))
print(new_abs(0))
```

__arguments and kwargs__

```python
def fun(*args):
    for arg in args:
        print(arg)
fun('hey','my','name','is','emon')
print('\n')
fun('age',21)
```

```python
def fun(args1,args2,*args):
    print('fix arguments 1: ',args1)
    print('fix arguments 1: ',args2)
    for arg in args:
        print(arg)
fun('hey','my','name','is','emon')
print('\n')
fun('age',21)
```

```python
def fun(arg,**kwargs):
    print('Data: ',arg)
    for key,val in kwargs.items():
        print(f'{key} -> {val}')
fun(fname = 'yash', lname='kumar',age=21)
```

__application__

```python
''' prime numbers '''
def isPrime(n):
    for i in range(2, int(1+n/2)):
        if n%i == 0:
            return False
        return True
print(isPrime(10))
print(isPrime(11))
```

```python
# add numbers from 1 - N
def add(n):
    ans = sum(range(1,n+1))
    return ans
print(add(10))
```

```python
# factorial of number n
def fac(n):
    ans = 1
    for i in range(1, n+1):
        ans *= i
        return ans
print(fac(4))
```

```python
import numpy as np
import matplotlib.pyplot as plt
def generate_data(n):
    e_values = []
    for i in range(n):
        e = np.random.randn()
        e_values.append(e)
    return e_values
data = generate_data(100)
plt.plot(data)
plt.show()
```
